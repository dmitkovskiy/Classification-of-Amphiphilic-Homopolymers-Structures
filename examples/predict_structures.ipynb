Load models, set error values

```python
model_phi_theta_path = "/content/Classification-of-Amphiphilic-Homopolymers-Structures/weights/model_phi_theta.pth"
model_phi_theta = mmf.load_my_image_pred_model(model_phi_theta_path, image_height=18, image_width=24)

model_x_y_path = "/content/Classification-of-Amphiphilic-Homopolymers-Structures/weights/model_x_y.pth"
model_x_y = mmf.load_my_image_pred_model(model_x_y_path, image_height=24, image_width=24)

models = {
    'model_phi_theta': model_phi_theta,
    'model_x_y': model_x_y
}

model_errors = {

    'model_phi_theta': {
        'Sector': 0., 'Part of helicoid': 0.05, 'Disk': 0.03,
        'Helicoid': 0, 'Enneper': 0, 'Complex structure': 0
    },

    'model_x_y': {
        'Sector': 0., 'Part of helicoid': 0.03, 'Disk': 0.05,
        'Helicoid': 0, 'Enneper': 0.05, 'Complex structure': 0.05
    }
}
```

## Predict Catenoid

Load structure point cloud

```python
main_path = "/content/Classification-of-Amphiphilic-Homopolymers-Structures/data/structures/"
structure_path = "catenoid_example.npy"
structure = np.load(main_path+structure_path)
```

Clustering the structure

```python
clustered_elements = prd.select_elements_sort_DBSCAN(structure)
clustered_dots = [el_dots for el_number, el_dots in clustered_elements]
prd.plot_points_3D(clustered_dots)
```

Demonstration of pixel projections in spherical *φ-θ* coordinates and cartesian *x-y* coordinates

```python
prd.plot_points_2D(clustered_dots)
```

Structure classification

```python
print(mmf.pred_structure(structure, models, model_errors)[0])
```
